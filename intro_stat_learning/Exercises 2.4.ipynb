```python
# Import libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('ticks')
```

```python
# Import and view the dataset.
df = pd.read_csv('college.csv')
df.rename(columns={ 'Unnamed: 0': 'Name' }, inplace=True)
df.set_index('Name', inplace=True)
df.head()
```

```python
# Show the summary.
df.describe()
```

```python
# Create a scatterplot matrix of the first 10 features.
sns.pairplot(df.iloc[:, 0:10])
plt.show()
```

```python
# Create boxplots our Outstate and Private.
sns.boxplot(df['Private'], df['Outstate'])
plt.show()
```

```python
# Create new 'elite' variable.
df['Elite'] = df['Top10perc'] > 50
df.head()
```

```python
# Create boxplot of 'outstate' vs 'elite'.
sns.boxplot(df['Elite'], df['Outstate'])
plt.show()
```

```python
# Plot some distributions.
f, axs = plt.subplots(2, 2)
sns.distplot(df['Top10perc'], ax=axs[0][0])
sns.distplot(df['Top25perc'], ax=axs[0][1])
sns.distplot(df['F.Undergrad'], ax=axs[1][0])
sns.distplot(df['P.Undergrad'], ax=axs[1][1])
plt.show()
```
