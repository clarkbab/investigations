### Law of Large Numbers

The law of large numbers states that as our sample size grows:
- The mean of the samples tends towards the population mean.
- The density distribution of the samples tends towards the density distribution of the population.

```python
n_samps = 100
s_small = np.random.choice(pop, n_samps)
s_small.size
```

```python
plt.hist(s_small, density=True)
plt.title('X Frequency Histogram')
plt.xlabel('X')
plt.ylabel('Num Occurences')
plt.show()
```

### Central Limit Theorem

The Central Limit Theorem states that as the sample size grows, the distribution of the average (or sum) of the samples will be normally distributed.

```python
n_draws = 1000
small_means = []
for i in range(n_draws):
    # For sample size of 1, this will have the same distribution as X.
    n_samps = 1
    s_small = np.random.choice(pop, n_samps, replace=True)
    small_means = np.append(small_means, s_small.mean())
```

```python
plt.hist(small_means, density=True)
plt.title('X Mean Density Histogram')
plt.xlabel('X')
plt.ylabel('Num Occurences')
plt.show()
```

```python
n_draws = 1000
medium_means = []
for i in range(n_draws):
    # For a sample size of 2, this will take a triangular distribution form.
    n_samps = 2
    s_medium = np.random.choice(pop, n_samps, replace=True)
    medium_means = np.append(medium_means, s_medium.mean())
```

```python
plt.hist(medium_means, density=True)
plt.title('X Mean Density Histogram')
plt.xlabel('X')
plt.ylabel('Num Occurences')
plt.show()
```

```python
n_draws = 1000
large_means = []
for i in range(n_draws):
    # With 3, we can already see it forming the shape of a bell curve.
    n_samps = 3
    s_large = np.random.choice(pop, n_samps, replace=True)
    large_means = np.append(large_means, s_large.mean())
```

```python
plt.hist(large_means, density=True)
plt.title('X Mean Density Histogram')
plt.xlabel('X')
plt.ylabel('Num Occurences')
plt.show()
```
